```python
import requests  # [+] HTTP 요청
from bs4 import BeautifulSoup  # [+] 웹스크래핑
import pandas as pd
```

```python
# 1. User-Agent 설정
# 우리가 실제 크롬 브라우저를 사용하는 일반 사용자인 것처럼 서버를 속이는 '위장 신분증' 역할
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

search_keyword = '이란'  # [+] 검색 키워드
url = f'https://search.naver.com/search.naver?where=news&query={search_keyword}'  # 검색한 웹페이지 주소
```

```python
# 2. 페이지 요청 및 파싱
res = requests.get(url, headers=headers)  # [+] 네이버 서버에 페이지 정보 요청(GET)
res.raise_for_status() # 서버가 응답을 잘 주었는지 확인(정상 200, 페이지 없음 404 등)
soup = BeautifulSoup(res.text, 'html.parser')  # [+] BeautifulSoup 객체 생성
```

```python
# 3. 질문자님이 찾은 클래스로 타겟팅 (띄어쓰기를 '.'으로 변경)
# [+] 찾고자 하는 HTML 태그 정의
target_class = "span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-2.sds-comps-text-type-headline1"
title_spans = soup.select(target_class) # [+] 조건에 맞는 모든 태그를 찾아 리스트로 저장

extracted_data = []
```

```python
extracted_data=[]
for span in title_spans:
    text_data = span.text.strip()
    extracted_data.append(text_data)
print(extracted_data)
```

```python
# 파이썬이 가져온 전체 글자 덩어리 출력해보기
print(soup.text)
```

```python
# 여러 클래스 중 가장 특징적인 것 하나만 사용해보기
target_class = "span.sds-comps-text-type-headline1"
```

```python
# 4. 데이터 추출
for span in title_spans:
    # [+] 기사 제목 추출
    title = span.text.strip
    
    # [+] 기사 링크 추출: 제목(span)을 클릭 가능하게 만드는 부모 <a> 태그를 역추적하여 찾음
    parent_a_tag = span.find_parent('a')
    link = parent_a_tag['href'] if parent_a_tag and parent_a_tag.has_attr('href') else "링크 없음"
    
    extracted_data.append({
        '제목': title,
        '링크': link
    })

df = pd.DataFrame(extracted_data)
print(f"\n총 {len(df)}개의 기사를 추출했습니다.\n")
display(df)
```

```python
import pandas as pd

# 1. 클래스명을 짧게 줄여서 다시 타겟팅해보기 (이전 단계에서 수정한 부분)
target_class = "span.sds-comps-text-type-headline1"
title_spans = soup.select(target_class) 

# [+] 태그를 제대로 찾았는지 먼저 확인!
print(f"찾아낸 태그 개수: {len(title_spans)}개")

# 2. 데이터 추출
extracted_data = []

for span in title_spans:
    # [+] 기사 제목 추출 (오타 수정: strip() 괄호 추가)
    title = span.text.strip()
    
    # [+] 기사 링크 추출
    parent_a_tag = span.find_parent('a')
    link = parent_a_tag['href'] if parent_a_tag and parent_a_tag.has_attr('href') else "링크 없음"
    
    extracted_data.append({
        '제목': title,
        '링크': link
    })

# 3. DataFrame으로 변환 및 확인
df = pd.DataFrame(extracted_data)
print(f"\n총 {len(df)}개의 기사를 추출했습니다.\n")
display(df)
```
